<font color="blue" size ="6"><b>Create a sample dataset and apply pre-processing techniques</b></font>

```python
import pandas as pd
import numpy as np

data = {
    "Age": [25, 30, np.nan, 45, 35],
    "Income": [50000, 60000, 52000, np.nan, 58000],
    "Gender": ["Male", "Female", "Female", "Male", "Other"],
    "Purchased": ["Yes", "No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)
print(df)
```

<font color="red" size="5" ><b>Pre-processing Techniques</font>

```python
#A. Handle Missing Values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Income"].fillna(df["Income"].mean(), inplace=True)
```

```python
#B. Encode Categorical Variables
#Label Encoding (for binary or yes/no)
df["Purchased"] = df["Purchased"].map({"Yes": 1, "No": 0})
```

```python
#One-Hot Encoding for Gender
df = pd.get_dummies(df, columns=["Gender"], drop_first=True)
```

```python
#C. Feature Scaling
#Standardization (Z-score scaling)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[["Age", "Income"]] = scaler.fit_transform(df[["Age", "Income"]])
```

```python
#D. Normalization (Optional)
from sklearn.preprocessing import MinMaxScaler

normalizer = MinMaxScaler()
df[["Age", "Income"]] = normalizer.fit_transform(df[["Age", "Income"]])
```
