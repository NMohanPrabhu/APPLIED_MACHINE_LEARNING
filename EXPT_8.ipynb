```python
# Sample dataset for EnjoySport
dataset = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High',   'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Cool', 'Change', 'Yes'],
    ['Sunny', 'Warm', 'Normal', 'Light',  'Warm', 'Same', 'Yes']
]

# Attribute names (optional, for clarity)
attributes = ['Sky', 'AirTemp', 'Humidity', 'Wind', 'Water', 'Forecast']
```

```python
def find_s_algorithm(dataset):
    # Step 1: Initialize hypothesis to most specific
    hypothesis = ["Ø"] * 6   # 6 attributes
    
    # Step 2: Iterate through training examples
    for example in dataset:
        attributes_values = example[:-1]  # All columns except target
        target = example[-1]
        
        # Consider only positive examples
        if target == 'Yes':
            # Update the hypothesis
            for i in range(len(hypothesis)):
                if hypothesis[i] == "Ø":
                    hypothesis[i] = attributes_values[i]
                elif hypothesis[i] != attributes_values[i]:
                    hypothesis[i] = "?"
                    
    return hypothesis
```

```python
final_hypothesis = find_s_algorithm(dataset)
print("Final Maximally Specific Hypothesis:")
print(final_hypothesis)
```
